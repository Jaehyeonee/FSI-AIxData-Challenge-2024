````python
# 범주형 데이터 원-핫 인코딩 수행
# 순서형 데이터 train['Customer_credit_rating'] 먼저 OrdinalEncoder 수행.
from sklearn.preprocessing import OrdinalEncoder

categories = [['S', 'A', 'B', 'C', 'D', 'E']]

# OrdinalEncoder 초기화 및 학습
encoder_train = OrdinalEncoder(categories=categories)
train['Customer_credit_rating'] = encoder_train.fit_transform(train[['Customer_credit_rating']])

# float형 변환 (선택사항)
train['Customer_credit_rating'] = train['Customer_credit_rating'].astype(float)



encoder_test= OrdinalEncoder(categories=categories)
test['Customer_credit_rating'] = encoder_test.fit_transform(test[['Customer_credit_rating']])

test['Customer_credit_rating'] = test['Customer_credit_rating'].astype(float)
````

````python
# 명목형 데이터 LabelEncoder 수행
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder

class MultiColLabelEncoder:
    def __init__(self):
        self.encoder_dict = defaultdict(LabelEncoder)

    def fit_transform(self, X: pd.DataFrame, columns : list):
        if not isinstance(columns, list):
            columns = [columns]

        output = X.copy()
        output[columns] = X[columns].apply(lambda x:self.encoder_dict[x.name].fit_transform(x))

        return output

mcle = MultiColLabelEncoder()
columns = ['Customer_Gender',
           'Customer_loan_type',
           'Account_account_type',
           'Channel',
           'Operating_System',
           'Error_Code',
           'Type_General_Automatic',
           'Access_Medium']
transformed_data_train = mcle.fit_transform(train, columns = columns)
transformed_data_train.head()

transformed_data_test = mcle.fit_transform(test, columns = columns)
transformed_data_test.head()

````
